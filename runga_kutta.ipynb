```python
import numpy as np
import matplotlib.pyplot as plt

# Define the interval and step size
dx = 0.01  # Fine step size for smoother plot
x = np.arange(-4, -2.8 + dx, dx)

# Compute the functions
y_exp = np.exp(x)
y_tan = np.tan(x)

# Create the plot
plt.plot(x, y_exp, label='e^x', color='blue')
plt.plot(x, y_tan, label='tan(x)', color='orange')
plt.axhline(0, color='black', linestyle='--')  # X-axis for reference
plt.grid(True)
```
