```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import solve_ivp
from scipy.special import factorial
```

```python
x=np.linspace(-10,10,2000)
y=np.sin(x)
plt.plot(x,y,'k',linewidth=2)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.grid(True)



```

```python
# first-order Taylor Series
P=[1,0]
yT1=np.polyval(P,x)
plt.plot(x,yT1,'b--',linewidth=1.2)
```

```python
# Third-order Taylor Series
P=[-1/factorial(3),0,1,0]
yT3=np.polyval(P,x)
plt.plot(x,yT3,'r--',linewidth=1.2)
```

```python
# Fifth-order Taylor Series
P=[1/factorial(5),0,-1/factorial(3),0,1,0]
yT5=np.polyval(P,x)
plt.plot(x,yT5,'g--',linewidth=1.2)
```

```python
# Sevent-order Taylor Series
P=[-1/factorial(7),0,1/factorial(5),0,-1/factorial(3),0,1,0]
yT7=np.polyval(P,x)
plt.plot(x,yT7,'m--',linewidth=1.2)
```

```python
# Ninth-order Taylor Series
P=[1/factorial(9),0,-1/factorial(7),0,1/factorial(5),0,-1/factorial(3),0,1,0]
yT9=np.polyval(P,x)
plt.plot(x,yT9,'c--',linewidth=1.2)
```

```python
x=np.linspace(-10,10,2000)
y=np.sin(x)
plt.plot(x,y,'k',linewidth=2)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.grid(True)

# first-order Taylor Series
P=[1,0]
yT1=np.polyval(P,x)
plt.plot(x,yT1,'b--',linewidth=1.2)

# Third-order Taylor Series
P=[-1/factorial(3),0,1,0]
yT3=np.polyval(P,x)
plt.plot(x,yT3,'r--',linewidth=1.2)

# Fifth-order Taylor Series
P=[1/factorial(5),0,-1/factorial(3),0,1,0]
yT5=np.polyval(P,x)
plt.plot(x,yT5,'g--',linewidth=1.2)

# Sevent-order Taylor Series
P=[-1/factorial(7),0,1/factorial(5),0,-1/factorial(3),0,1,0]
yT7=np.polyval(P,x)
plt.plot(x,yT7,'m--',linewidth=1.2)

# Ninth-order Taylor Series
P=[1/factorial(9),0,-1/factorial(7),0,1/factorial(5),0,-1/factorial(3),0,1,0]
yT9=np.polyval(P,x)
plt.plot(x,yT9,'c--',linewidth=1.2)

plt.legend(["sin(x)","1st order","3rd order","5th order","7th order","9th order"])
plt.show()
```
